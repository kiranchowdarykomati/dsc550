```python
import pandas as pd
df_mrd=pd.read_csv('labeledTrainData.tsv',sep='\t')
```

```python
# Convert all text to lowercase letters.
df_mrd['review'] = df_mrd['review'].str.lower()
df_mrd.head()
```

```python
# Remove punctuation and special characters from the text.
import unicodedata
import sys

punctuation = dict.fromkeys(i for i in range(sys.maxunicode)
if unicodedata.category(chr(i)).startswith('P'))
# For each string, remove any punctuation characters
df_mrd['review'] = df_mrd['review'].apply(lambda x: x.translate(punctuation))
df_mrd.head()
```

```python
# Remove stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
df_mrd['review_without_stopwords'] = df_mrd['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df_mrd.head()
```

```python
# Apply porter stemmer
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
# Tokenize
df_mrd['review_tokens'] = df_mrd['review_without_stopwords'].apply(lambda x: word_tokenize(x))
# Stem
df_mrd['review_stem'] = df_mrd['review_tokens'].apply(lambda x : [ps.stem(y) for y in x])
```

```python
df_mrd.head()
```

```python
# bag of words matrix
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
#feature = np.array(df_mrd['review_stem'])
vectorizer = CountVectorizer()
word_matrix=df_mrd['review_stem'].apply(lambda x: vectorizer.fit_transform(x))
len(word_matrix)
```

```python
# tf-idf matrix
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
#feature_matrix = tfidf.fit_transform(text_data)
word_matrix=df_mrd['review_stem'].apply(lambda x: tfidf.fit_transform(x))
# Show tf-idf feature matrix
len(word_matrix)
```
